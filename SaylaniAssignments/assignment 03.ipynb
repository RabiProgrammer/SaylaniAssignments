3) Find and display the largest number of a list without using built_in function max(). Your program should ask the user to input values in list from keyboard.

```python
user = input("Enter numbers separated by spaces: ")

lists = [int(x) for x in user.split()]

largest = lists[0]

for num in lists:
    if num > largest:
        largest = num

print("The largest number is: ", largest)
```

4) Write a program that rotates the element of a list so that the element at the first index moves to the second index, the element in the second index moves to the third index, etc. and the element in the last index moves to the first index.

```python
user_input = input("Enter elements of the list separated by spaces: ")

lists = user_input.split()

last_element = lists[-1]

for i in range(len(lists)-1, 0, -1):
    lists[i] = lists[i-1]

lists[0] = last_element

print("Rotated list:", lists)
```

5) Write a program that input a string and ask the user to delete a given word from a string.

```python
string = input("Enter a string: ")

word_del = input("Enter a word you want to delete: ")

new_str = ' '.join([word for word in string.split() if word != word_del])

print("Updated string: ", new_str)
```

6) Write a program that reads a string from the user containing a date in the form mm/dd/yyyy. It should print the date in the form March 12, 2021.

```python
date_input = input("Enter a date in mm/dd/yyyy format: ")

month, day, year = date_input.split('/')

month_names = {
    "01": "January", "02": "February", "03": "March", "04": "April",
    "05": "May", "06": "June", "07": "July", "08": "August",
    "09": "September", "10": "October", "11": "November", "12": "December"
}

month_name = month_names[month]

print(f"{month_name} {int(day)}, {year}")
```

7) Write a program with a function that accepts a string from keyboard and create a new string after converting character of each word capatilized. For instance, if the sentence is "stop and smell the roses." the output should be "Stop And Smell The Roses"

```python
user_input = input("Enter a sentence: ")

words = user_input.split()

capitalized_words = []
for word in words:
    capitalized_word = word[0].upper() + word[1:]
    capitalized_words.append(capitalized_word)

new_sentence = ' '.join(capitalized_words)

print("Capitalized sentence:", new_sentence)
```

8) Find the sum of each row of matrix of size mxn. For example for the following matrix output will be like this:
2 11 7 12
5 2  9 15
8 3 10 42
Sum of row 1 = 32
Sum of row 2 = 31
Sum of row 3 = 63

```python
m = int(input("Enter number of rows: "))
n = int(input("Enter number of columns: "))

matrix = []

print("Enter matrix elements row-wise:")
for i in range(m):
    row = []
    for j in range(n):
        row.append(int(input()))
    matrix.append(row)

print("\nRow sums:")
for i in range(m):
    print(f"Sum of row {i+1} =", sum(matrix[i]))
```

9) Write a program to add two matrices of size n x m.

```python
n = int(input("Enter number of rows (n): "))
m = int(input("Enter number of columns (m): "))

print("\nEnter elements of first matrix:")
A = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(int(input()))
    A.append(row)

print("\nEnter elements of second matrix:")
B = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(int(input()))
    B.append(row)

C = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(A[i][j] + B[i][j])
    C.append(row)

print("\nResultant matrix after addition:")
for row in C:
    print(row)
```

10) Write a program to multiply two matrices.

```python
n = int(input("Enter number of rows (n): "))
m = int(input("Enter number of columns (m): "))

print("\nEnter elements of first matrix:")
A = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(int(input()))
    A.append(row)

print("\nEnter elements of second matrix:")
B = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(int(input()))
    B.append(row)

C = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(A[i][j] * B[i][j])
    C.append(row)

print("\nResultant matrix after multiplication:")
for row in C:
    print(row)
```
