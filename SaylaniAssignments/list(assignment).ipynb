Q1: Write a program that accepts a list from user and print the alternate element of list.

```python
lst = input("Enter elements of the list separated by space: ").split()

print("Alternate elements of the list: ")
for i in range(0, len(lst), 2):
    print(lst[i])
```

Q2: Write a program that accepts a list from user. Your program should reverse the content of list and display it. Do not use reverse() method.

```python
lst = input("Enter elements of the list separated by space: ").split()

reversed_list = lst[::-1]
print("Reversed list: ", reversed_list)
```

Q3: Find and display the largest number of a list without using built-in function max(). Your program should ask the user to input values in list from keyboard.

```python
lst = list(map(int, input("Enter elements of the list separated by space: ").split()))

largest = lst[0]

for num in lst:
    if num > largest:
        largest = num

print("Largest element in the list: ", largest)
```

Q6. Write a program that reads a string from the user containing a date in the form mm/dd/yyyy. It should print the date in the form March 12, 2021.

```python
date = input("Enter date in mm/dd/yyyy format: ")

mm, dd, yyyy = date.split("/")

months = {
    "01": "January",
    "02": "February",
    "03": "March",
    "04": "April",
    "05": "May",
    "06": "June",
    "07": "July",
    "08": "August",
    "09": "September",
    "10": "October",
    "11": "November",
    "12": "December"
}

print(f"{months[mm]} {int(dd)} {yyyy}")
```

Q8. Find the sum of each row of matrix of size m x n. For example for the following matrix output will be like this :

[2, 11, 7, 12]
[5, 2, 9, 15]
[8, 3, 10, 42]
Sum of row 1 = 32
Sum of row 2 = 31
Sum of row 3 = 63

```python
lst = [
    [2, 11, 7, 12],
    [5, 2, 9, 15],
    [8, 3, 10, 42]
]

count = 0

for x in range(len(lst)):
    sum = 0
    for y in range(len(lst[x])):
        sum += lst[x][y]
    print(f"Sum of list {count} = {sum}")
    count += 1
```

Q9: Write a program to add two matrices of size n x m.

```python
lst = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

lst2 = [
    [10, 11, 12],
    [13, 14, 15],
    [16, 17, 18]
]

result = []

for i in range(len(lst)):
    row = []
    for j in range(len(lst[i])):
        row.append(lst[i][j] + lst2[i][j])
    result.append(row)

print("Result of addition:")
for r in result:
    print(r)
```
