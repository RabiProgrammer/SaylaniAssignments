```python
# section_a a3

# Input from user
price = float(input("Enter item price: "))
quantity = int(input("Enter quantity: "))
discount_percent = float(input("Enter discount %: "))
tax_rate = float(input("Enter tax rate %: "))

# Subtotal
subtotal = price * quantity

# Discount amount
discount_amount = (discount_percent / 100) * subtotal

# Amount after discount
after_discount = subtotal - discount_amount

# Tax amount
tax_amount = (tax_rate / 100) * after_discount

# Grand total (rounded)
grand_total = round(after_discount + tax_amount, 2)

# Output
print("Subtotal:", subtotal)
print("Discount amount:", discount_amount)
print("After discount:", after_discount)
print("Tax amount:", tax_amount)
print("Grand total:", grand_total)
```

```python
# section b b_3

count = 0
total = 0
even_count = 0
odd_count = 0
largest = None
smallest = None

while True:
    user_input = input("Enter a number (or 'done' to stop): ")

    # Stop condition
    if user_input.lower() == "done":
        break

    # Handle invalid input
    if not user_input.isdigit():
        print("Warning: Invalid input skipped (not a number)")
        continue

    num = int(user_input)

    # Update counters
    count += 1
    total += num

    # Even / Odd check
    if num % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

    # Largest / Smallest
    if largest is None or num > largest:
        largest = num

    if smallest is None or num < smallest:
        smallest = num

# Final results
if count == 0:
    print("No valid numbers entered.")
else:
    average = round(total / count, 2)

    print("\n--- Results ---")
    print("Total numbers:", count)
    print("Sum:", total)
    print("Average:", average)
    print("Largest:", largest)
    print("Smallest:", smallest)
    print("Even count:", even_count)
    print("Odd count:", odd_count)
```

```python
# section c c_1

# Check if number is prime
def is_prime(n):
    """Returns True if n is a prime number, otherwise False."""
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True


# Celsius to Fahrenheit and Kelvin
def celsius_to_all(c):
    """Converts Celsius to Fahrenheit and Kelvin and returns a tuple."""
    fahrenheit = (c * 9/5) + 32
    kelvin = c + 273.15
    return fahrenheit, kelvin


# Count vowels in a string
def count_vowels(text):
    """Returns the number of vowels in a given string."""
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count


# Power function
def power(base, exp=2):
    """Returns base raised to the power of exp (default is square)."""
    return base ** exp

print("Is 7 prime?", is_prime(7))
print("Is 10 prime?", is_prime(10))

print("\nCelsius 25 ->", celsius_to_all(25))

print("\nVowel count:", count_vowels("Hello World"))

print("\nPower 3^2:", power(3))
print("Power 2^3:", power(2, 3))
```

```python
# section c c_2

# Function 1: Report Card Generator
def generate_report(**student_info):
    """Generates and prints a formatted student report card."""

    name = student_info.get("name")
    roll_no = student_info.get("roll_no")
    marks = student_info.get("marks", [])
    subjects = student_info.get("subjects", [])

    # Calculate average
    average = sum(marks) / len(marks)

    # Grade logic
    if 90 <= average <= 100:
        grade = "A+"
    elif 80 <= average < 90:
        grade = "A"
    elif 70 <= average < 80:
        grade = "B"
    elif 60 <= average < 70:
        grade = "C"
    elif 50 <= average < 60:
        grade = "D"
    else:
        grade = "F"

    # Print report
    print("----- Report Card -----")
    print("Name:", name)
    print("Roll No:", roll_no)
    print("Subjects:", subjects)
    print("Marks:", marks)
    print("Average:", round(average, 2))
    print("Grade:", grade)


# Function 2: Sum of any number of values
def total_marks(*args):
    """Returns sum of all given numbers."""
    return sum(args)

generate_report(
    name="Sana Khan",
    roll_no="CS-101",
    subjects=["Math", "Physics", "Python"],
    marks=[85, 78, 92]
)

print("\nTotal Marks:", total_marks(85, 78, 92))
```

```python
# section c c_4 part 1

def fibonacci(n):
    """Returns nth Fibonacci number using recursion."""
    if n == 0:
        return 0
    elif n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)


# Print first 10 Fibonacci numbers
print('Part 1')
print("\n\nFibonacci Series:")
for i in range(10):
    print(fibonacci(i), end=" ")

# section c c_4 part 2

def factorial(n):
    """Returns factorial of n using recursion. Handles negative input."""
    if n < 0:
        return "Invalid input (negative number)"
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)


# Test cases
print('\n\nPart 2')
print("\nFactorial of 5:", factorial(5))
print("Factorial of 0:", factorial(0))
print("Factorial of -3:", factorial(-3))

# section c c_4 part 3

def flatten(lst):
    """Flattens a deeply nested list using recursion."""
    result = []

    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)

    return result


# Test case
nested = [1, [2, [3, 4], 5], [6, 7]]
print('\n\nPart 3')
print("\nFlattened List:", flatten(nested))
```

```python
# section d d_1

inventory = {
    "Laptop":     {"price": 85000, "qty": 10, "category": "Electronics"},
    "T-Shirt":    {"price": 1200,  "qty": 50, "category": "Clothing"},
    "Headphones": {"price": 5500,  "qty": 20, "category": "Electronics"},
    "Jeans":      {"price": 3200,  "qty": 30, "category": "Clothing"},
    "Mouse":      {"price": 1800,  "qty": 15, "category": "Electronics"},
}

# Total inventory value
total_value = sum(details["price"] * details["qty"] for details in inventory.values())
print("Total Inventory Value:", total_value)


# Most expensive item
most_expensive = max(inventory.items(), key=lambda x: x[1]["price"])
print("\nMost Expensive Item:", most_expensive[0], "-", most_expensive[1]["price"])


# Electronics items
electronics = [name for name, details in inventory.items() if details["category"] == "Electronics"]
print("\nElectronics Items:", electronics)


# Apply 10% discount if price > 5000
for item in inventory:
    if inventory[item]["price"] > 5000:
        inventory[item]["price"] *= 0.9  # 10% discount

print("\nInventory after discount:")
print(inventory)


# Sorted list by price (ascending)
sorted_items = sorted(inventory.items(), key=lambda x: x[1]["price"])

print("\nItems sorted by price (ascending):")
for item in sorted_items:
    print(item[0], "-", item[1]["price"])
```

```python
# section d d_3

class Book:
    """Represents a book in the library."""

    def __init__(self, title, author, isbn, is_available=True):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_available = is_available

    def __str__(self):
        status = "Available" if self.is_available else "Checked Out"
        return f"{self.title} by {self.author} (ISBN: {self.isbn}) - {status}"

    def checkout(self):
        """Marks book as not available."""
        if self.is_available:
            self.is_available = False
        else:
            print(f"{self.title} is already checked out.")

    def return_book(self):
        """Marks book as available."""
        self.is_available = True


class Library:
    """Manages a collection of books."""

    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def find_by_author(self, author):
        return [book for book in self.books if book.author == author]

    def available_books(self):
        return [book for book in self.books if book.is_available]

    def total_books(self):
        return len(self.books)


book1 = Book("Python Basics", "Rabi Khan", "101")
book2 = Book("AI Fundamentals", "Qasim Hussain", "102")
book3 = Book("Data Science", "Alina Imran", "103")
book4 = Book("Web Development", "Zain Mukthar", "104")

library = Library()

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)
library.add_book(book4)

book1.checkout()
book2.checkout()

print("Available Books:")
for book in library.available_books():
    print(book)

print("\nTotal Books:", library.total_books())

print("\nBooks by Qasim Hussain:")
for book in library.find_by_author("Qasim Hussain"):
    print(book)
```

```python
# section d d_4

def save_student(name, marks):
    """Appends a student record to file."""
    with open("students.txt", "a") as f:
        f.write(f"{name},{marks}\n")


def load_students():
    """Reads file and returns list of student dictionaries."""
    students = []

    try:
        with open("students.txt", "r") as f:
            for line in f:
                name, marks = line.strip().split(",")
                students.append({
                    "name": name,
                    "marks": int(marks)
                })
    except FileNotFoundError:
        print("File not found. No students available.")

    return students


def top_student():
    """Returns name of student with highest marks."""
    students = load_students()

    if not students:
        return None

    top = max(students, key=lambda s: s["marks"])
    return top["name"], top["marks"]

save_student("Ali", 85)
save_student("Sara", 92)
save_student("Umar", 78)
save_student("Hina", 88)
save_student("Bilal", 91)


result = top_student()

if result:
    print("Top Student:", result[0], "with marks:", result[1])
```
